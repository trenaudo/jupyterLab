```python
# import libraries to use (pandas, matplotlib.pyplot and seaborn)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv')
df
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
df
```

```python
data_types = df.dtypes
data_types
```

```python
df = df.drop("unnamed:_0", axis=1)
df
```

```python
df = df.dropna() #null values 
df
```

```python
df['effective_to_date'] = pd.to_datetime(df['effective_to_date'])

df['month'] = df['effective_to_date'].dt.month

df
```

1- Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.


```python
new_df = df[(df['total_claim_amount'] > 1000) & (df['response'] == 'Yes')]
new_df
```

2. Analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign.

```python
df_yes = df[df['response'] == 'Yes']
df_yes
```

```python
df_yes.groupby(['policy_type', 'gender'])['total_claim_amount'].mean()
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
df_state = df.groupby(['state'])['customer'].count()
df_state
```

```python
filtered_states = df_state[df_state > 500]
filtered_states
```

4. Find the median customer lifetime value by education level and gender.

```python
df_clv = df.groupby(['education','gender'], as_index=False)['customer_lifetime_value'].median() #as_index=False to display as a table
df_clv

#as_index=False to display as a table
```

5. Find the maximum, minimum, and average number of open complaints by state and policy type.

```python
complaints = df.groupby(['state', 'policy_type'])['number_of_open_complaints'].agg(['max', 'min', 'mean'])
complaints
```

6. The marketing team wants to analyze the number of policies sold by state and month.

Present the data in a table where the months are arranged as columns and the states are arranged as rows.

```python
pivot_table = pd.pivot_table(df, values='policy', index='state', columns='month', aggfunc='count')
pivot_table
```

7. Display a new DataFrame that contains the number of policies sold by month by state for the top 3 states with the highest number of policies sold.

*Hint:* 
- *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.*
- *Next, you will select the top 3 states with the highest number of policies sold.*
- *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*

```python
group = df.groupby(['state', 'month'])['policy'].count().reset_index()
group
```

```python
sorted_data = group.sort_values(by='policy', ascending=False)
sorted_data
```

```python
top_states = sorted_data['state'].unique()[:3]
top_states

```

```python
filtered_data = sorted_data[sorted_data['state'].isin(top_states)]
filtered_data
```

```python
pivot_table = pd.pivot_table(filtered_data, values='policy', index='state', columns='month', aggfunc='sum', fill_value=0)
pivot_table
```
