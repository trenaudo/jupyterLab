```python
# import libraries to use (pandas, matplotlib.pyplot and seaborn)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dfr3 = pd.read_csv('https://raw.githubusercontent.com/repicao/IH_AB_DA_FT_MAY_2023/master/Class_Materials/Case_Studies/Customer_Analysis_Case_Study/Data/Data_Marketing_Customer_Analysis_Round3.csv')
dfr3
```

```python
sns.countplot(x="response",data=dfr3,palette="Greens_d")

plt.xlabel('Response')
plt.ylabel('Total Responses')
plt.title('Total Number of Responses')

```

```python
sns.countplot(x="response", hue="sales_channel", data=dfr3, palette="Set2")

# labels and title
plt.xlabel('Response')
plt.ylabel('Count')
plt.title('Response by Sales Channel')

```

```python
sns.histplot(data=dfr3, x="total_claim_amount", hue="response")

# Adding labels and title
plt.xlabel('Total Claim Amount')
plt.ylabel('Count')
plt.title('Histogram of Response by Total Claim Amount')

# Displaying the plot
plt.show()
```

```python
# Filter the data for response == "yes"
data_yes = dfr3[dfr3["response"] == "yes"]

# Filter the data for response == "no"
data_no = dfr3[dfr3["response"] == "no"]

# Creating two histogram plots using seaborn
plt.figure(figsize=(12, 5))  # Adjust the figure size if needed

plt.subplot(1, 2, 1)
sns.histplot(data=data_yes, x="total_claim_amount", color="green")
plt.xlabel('Total Claim Amount')
plt.ylabel('Count')
plt.title('Histogram of Response: Yes')

plt.subplot(1, 2, 2)
sns.histplot(data=data_no, x="total_claim_amount", color="red")
plt.xlabel('Total Claim Amount')
plt.ylabel('Count')
plt.title('Histogram of Response: No')


```

```python
# Filter the data for response == "yes"
data_yes = dfr3[dfr3["response"] == "yes"]

# Filter the data for response == "no"
data_no = dfr3[dfr3["response"] == "no"]

# Creating two histogram plots using seaborn
plt.figure(figsize=(12, 5))  # Adjust the figure size if needed

plt.subplot(1, 2, 1)
sns.histplot(data=data_yes, x="income", color="green")
plt.xlabel('Income')
plt.ylabel('Count')
plt.title('Histogram of Response: Yes')

plt.subplot(1, 2, 2)
sns.histplot(data=data_no, x="income", color="red")
plt.xlabel('Income')
plt.ylabel('Count')
plt.title('Histogram of Response: No')
```
